# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self,model,speed,color):
        self.model=model
        self.speed=speed
        self.color=color
        
class Bus(Vehicle):
    def __init__(self,model,speed,color,capacity=50):
        super().__init__(model,speed,color)
        self.capacity=capacity
        print(f"model:{self.model},speed:{self.speed},color:{self.color},capacity:{self.capacity}")

bus1=Bus("Man",200,"Black",70)
````
